```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import Birch
from sklearn import metrics
```

# Dim dataset 

```python
df2 = pd.read_csv('dim2.txt', delim_whitespace=True)
df2_half=df2.loc[:674]

df5 = pd.read_csv('dim5.txt', delim_whitespace=True)
df5_half=df5.loc[:1687]

df10 = pd.read_csv('dim10.txt', delim_whitespace=True)
df10_half=df10.loc[:3374]

df15 = pd.read_csv('dim15.txt', delim_whitespace=True)
df15_half=df15.loc[:6749]
```

### Using:
df2, df5, df10, df15, df2_half, df5_half, df10_half, df15_half.

```python
df = pd.read_csv('dim2.txt', delim_whitespace=True)
X=np.asarray(df)

brc = Birch(branching_factor=50, n_clusters=10, threshold=0.5,compute_labels=True)

labels = brc.fit(X).predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis');
```
